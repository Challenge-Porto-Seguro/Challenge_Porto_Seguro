```python
def diagnostico_problema(problema):
    resultado = base_conhecimento[base_conhecimento["problema"] == problema]
    
    if not resultado.empty:
        print(f"Problema: {resultado['problema'].values[0]}")
        print("\nCausas possíveis:")
        for causa in resultado["causas"].values[0]:
            print(f"- {causa}")
        
        print("\nSoluções recomendadas:")
        for solucao in resultado["soluções"].values[0]:
            print(f"- {solucao}")
    else:
        print("Problema não encontrado na base de conhecimento.")


# Exemplo de uso
diagnostico_problema("motor não liga")
```

```python
def consulta_por_palavra_chave(descricao):
    # Busca problemas que contenham a descrição dada pelo usuário
    resultados = base_conhecimento[base_conhecimento["problema"].str.contains(descricao, case=False, na=False)]
    
    if not resultados.empty:
        for _, row in resultados.iterrows():
            print(f"\nProblema: {row['problema']}")
            print("Possíveis causas:")
            for causa in row["causas"]:
                print(f"- {causa}")
            print("Soluções recomendadas:")
            for solucao in row["soluções"]:
                print(f"- {solucao}")
    else:
        print("Nenhum problema encontrado com essa descrição.")
        
# Exemplo de uso
consulta_por_palavra_chave("barulho")
```

```python
def calcular_custo_estimado(problema):
    resultado = base_conhecimento[base_conhecimento["problema"] == problema]
    
    if not resultado.empty:
        custo_estimado = resultado["custo_estimado"].values[0]
        print(f"Custo estimado para resolver '{problema}': R$ {custo_estimado:.2f}")
    else:
        print("Problema não encontrado na base de conhecimento.")

calcular_custo_estimado("motor não liga")
```
